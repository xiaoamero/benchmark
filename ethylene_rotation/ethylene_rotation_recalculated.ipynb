```python
# ground state
import numpy as np
from pyscf import gto,scf,cc,fci,mcscf,mp,mrpt

ethylene = [ethylene0,ethylene10,ethylene20,ethylene30,ethylene40,
            ethylene50,ethylene55,ethylene60,ethylene65,ethylene70,
           ethylene75,ethylene80,ethylene83,ethylene85,ethylene88,ethylene90]
angle = [0,10,20,30,40,50,55,60,65,70,75,80,83,85,88,90]
angle_list=[]
fci_list=[]


for R in range(0,16):
    
    mol = gto.M(atom = ethylene[R], basis = 'sto-3g', verbose=3)
    dihedralangle = int(angle[R])
    angle_list.append(dihedralangle)
 #   print("angle = ",dihedralangle)
    
    molRHF = scf.RHF(mol)
    energy_RHF = molRHF.kernel()
 #   rhf_list.append(energy_RHF)
 
    molFCI = fci.FCI(mol, molRHF.mo_coeff)
    molFCI.nroots = 2
     #molFCI = fci.addons.fix_spin_(molFCI, ss=2) # S*(S+1)
    energy_FCI = molFCI.kernel()[0][0]
    fci_list.append(energy_FCI)
```
